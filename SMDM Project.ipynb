# Outlier Treatment

```python
df.columns
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range    
```

```python
lrt_sal,urt_sal=remove_outlier(df['Total_salary'])
df['Total_salary']=np.where(df['Total_salary']>urt_sal,urt_sal,df['Total_salary'])
df['Total_salary']=np.where(df['Total_salary']<lrt_sal,lrt_sal,df['Total_salary'])
```
