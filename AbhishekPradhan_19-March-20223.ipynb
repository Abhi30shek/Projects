## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

```python
#df['Newcolum'] = df['discounted_price'].apply(lambda x : x * .25 if x >600 )
    
    
```

```python
def charg(value):
    x = int(value)
    print(x)
    if x>600:
        y=x*.25
    elif x>350:
        y=x*.15
    elif x>100:
        y=x*.10
    else:
        y=x*0.5
    return y
```

```python
df['Revenue_calculated'] = 0
```

```python
for i in df.index:
    df.loc[i,'Revenue_calculated'] = charg(df.loc[i,'discounted_price'])
        
```

```python
df['Revenue_calculated'].sum()
        
```
